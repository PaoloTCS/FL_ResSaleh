```python
# Setting the number of points and the randomness of the points
np.random.seed(0) # for repeatability of this tutorial
npnts = 20 # number of points
pnts = np.random.rand(npnts,2)
```

```python
# Generate Voronoi Tessellation
vor = Voronoi(pnts)
```

```python
# Plot Voronoi diagram
voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=2)

plt.show()
```

```python
# Required Libraries
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Setting the number of points and the randomness of the points
np.random.seed(0) # for repeatability of this tutorial
npnts = 20 # number of points
pnts = np.random.rand(npnts,2) # generate 2D array of points
```

```python
# Generate Voronoi Tessellation
vor = Voronoi(pnts)

# Generate Delaunay Tessellation
deln = Delaunay(pnts)
```

```python
# Plot Voronoi diagram
voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=2)
```

```python
# Plot Delaunay edges
plt.triplot(pnts[:,0], pnts[:,1], deln.simplices.copy(), 'k--')

plt.show()
```

```python
# Required Libraries
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay 
import matplotlib.pyplot as plt
import numpy as np

# Setting the number of points and the randomness of the points
np.random.seed(0) # for repeatability of this tutorial
npnts = 20 # number of points
pnts = np.random.rand(npnts,2) # generate 2D array of points

# Generate Voronoi Tessellation
vor = Voronoi(pnts)

# Generate Delaunay Tessellation
deln = Delaunay(pnts)

# Plot Voronoi diagram
voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=2)

# Plot Delaunay edges
plt.triplot(pnts[:,0], pnts[:,1], deln.simplices.copy(), 'k--')

plt.show()
```
