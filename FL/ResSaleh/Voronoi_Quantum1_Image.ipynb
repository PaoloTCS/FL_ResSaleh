```python
# Re-import necessary modules and re-run the code to label the bounded Voronoi cells

import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt

# Create random points for Voronoi diagram (for demonstration)
np.random.seed(0)
points = np.random.rand(10, 2)

# Compute Voronoi diagram
vor = Voronoi(points)

# Plot Voronoi diagram
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)
ax.scatter(points[:, 0], points[:, 1], c='red', label='Voronoi Points')

# Label the bounded Voronoi cells
for i, region in enumerate(vor.regions):
    if not -1 in region and len(region) > 0:
        polygon = [vor.vertices[i] for i in region]
        ax.annotate(f'Cell {i}', np.mean(polygon, axis=0), color='blue', fontsize=12)

ax.set_title('Example Voronoi Diagram with Labeled Cells')
ax.legend()

plt.show()
```

```python
# Plot Voronoi diagram again with specific labeling
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)
ax.scatter(points[:, 0], points[:, 1], c='red', label='Voronoi Points')

# Label the Voronoi cell generators
for i, point in enumerate(points):
    ax.annotate(f'Generator {i}', point, color='black', fontsize=10, ha='right', va='bottom')

# Label all cells around cell 7
cell_7_neighbors = set()
for i, ridge_vertices in enumerate(vor.ridge_vertices):
    if 7 in vor.ridge_points[i]:
        cell_7_neighbors.add(vor.ridge_points[i][0] if vor.ridge_points[i][1] == 7 else vor.ridge_points[i][1])

for neighbor in cell_7_neighbors:
    polygon = vor.vertices[vor.regions[vor.point_region[neighbor]]]
    if all(vertex >= 0 for vertex in vor.regions[vor.point_region[neighbor]]):
        ax.annotate(f'Neighbor Cell {neighbor}', np.mean(polygon, axis=0), color='green', fontsize=12)

# Label Cell 7
polygon_7 = vor.vertices[vor.regions[vor.point_region[7]]]
ax.annotate(f'Cell 7', np.mean(polygon_7, axis=0), color='blue', fontsize=12)

ax.set_title('Example Voronoi Diagram with Labeled Cells and Generators')
ax.legend()

plt.show()
```

```python
# Re-running the code snippet to calculate the angle between two Delaunay edges

import numpy as np

def calculate_angle(A, B, C):
    # Convert points to numpy arrays for easier manipulation
    A, B, C = np.array(A), np.array(B), np.array(C)
    
    # Create vectors AB and BC
    AB = B - A
    BC = C - B
    
    # Calculate the dot product
    dot_product = np.dot(AB, BC)
    
    # Calculate the magnitudes of AB and BC
    magnitude_AB = np.linalg.norm(AB)
    magnitude_BC = np.linalg.norm(BC)
    
    # Calculate the cosine of the angle
    cos_theta = dot_product / (magnitude_AB * magnitude_BC)
    
    # Calculate the angle in radians
    theta_rad = np.arccos(np.clip(cos_theta, -1.0, 1.0))
    
    # Convert to degrees
    theta_deg = np.degrees(theta_rad)
    
    return theta_rad, theta_deg

# Sample points representing Generators 0, 1 and 5
A = np.array([0.5, 0.7])  # Coordinates for Generator 0
B = np.array([0.2, 0.4])  # Coordinates for Generator 1
C = np.array([0.8, 0.3])  # Coordinates for Generator 5

# Calculate the angle
theta_rad, theta_deg = calculate_angle(A, B, C)
theta_rad, theta_deg
# From Oct 29 session with ChatGPT 

```
