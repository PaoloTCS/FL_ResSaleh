```python
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay
import numpy as np

# Generate 10 random 2D points
points = np.random.rand(10, 2)

# Create Voronoi diagram
vor = Voronoi(points)

# Create Delaunay Triangulation
delaunay = Delaunay(points)

# Plot Voronoi diagram with blue lines
voronoi_plot_2d(vor, ax=None, show_vertices=False, line_colors='blue', line_width=2)

# Plot Delaunay edges in red
plt.triplot(points[:, 0], points[:, 1], delaunay.simplices, color='red', linewidth=1.0)

# Label points with capital letters
for i, point in enumerate(points):
    plt.text(point[0], point[1], chr(65 + i), ha='right')  # 65 is the ASCII value of 'A'

# Color each Voronoi region
for region in vor.regions:
    if not -1 in region:
        polygon = [vor.vertices[i] for i in region]
        plt.fill(*zip(*polygon), alpha=0.4)

plt.title("Figure 2: Voronoi Diagram and Delaunay Triangulation")
plt.show()
```
