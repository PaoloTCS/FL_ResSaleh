```python
# Required Libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay

# Generate 10 random 2D points
points = np.random.rand(10, 2)

# Create Voronoi diagram
vor = Voronoi(points)

# Create Delaunay triangulation
tri = Delaunay(points)

# Plot Voronoi diagram
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)

# Plot Delaunay triangulation on top of it
ax.triplot(points[:,0], points[:,1], tri.simplices, color='r')

# Label points with numbers for reference
for i, point in enumerate(points):
    plt.text(point[0], point[1], str(i), ha='right')

plt.title('Figure 2: Voronoi and Corresponding Delaunay Triangulation')
plt.show()
```
