```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Function to generate prime numbers
def generate_primes(n):
    primes = []
    i = 2
    while len(primes) < n:
        for p in primes:
            if i % p == 0:
                break
        else:
            primes.append(i)
        i += 1
    return primes

# Generate two-dimensional list of random integers (50x2)
var_rand = np.random.randint(0, 100, size=(50, 2))

# Create Voronoi tessellation using Fortune's algorithm (Scipy uses it)
vor = Voronoi(var_rand)

# Generate 50 prime numbers for the 50 cells
primes = generate_primes(50)

# Plot Voronoi tessellation
fig, ax = plt.subplots()
voronoi_plot_2d(vor, ax=ax)
ax.scatter(var_rand[:, 0], var_rand[:, 1], c='red', marker='o')

# Label Voronoi cells with ascending prime numbers
for i, (x, y) in enumerate(var_rand):
    ax.text(x, y, str(primes[i]), ha='center', va='center')

# Show plot
plt.show()
```
