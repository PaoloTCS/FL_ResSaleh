```python
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
import numpy as np

# Function to create and plot Voronoi diagram
def create_voronoi_fig1():
    # Generate 20 random 2D points
    points = np.random.rand(20, 2)

    # Create Voronoi diagram
    vor = Voronoi(points)

    # Plot Voronoi diagram
    voronoi_plot_2d(vor)
    
    # Label points with numbers
    for i, point in enumerate(points):
        plt.text(point[0], point[1], str(i+1), ha='right')
        
    plt.title("Fig1: Voronoi Diagram with 20 random points")
    plt.show()

# Generate Fig1
create_voronoi_fig1()
```
